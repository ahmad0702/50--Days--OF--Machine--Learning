# **Import Libraries**

```python
import pandas as pd
import numpy as np
# import matplotlib.pyplot as plt
# import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# **Loading Dataset .....**

```python
sonar_data = pd.read_csv('/content/Sonaar.csv', header = None)
```

# **Data Pre-Processing**

### **Drop Target Coulmn**

```python
X = sonar_data.drop(columns = 60, axis = 1)
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

### **Data Splitting**
Into Train And Test

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

# **Model Selection And Loading**

```python
model = LogisticRegression()
```

### **Fit Training And Testing Data**
In The Selected Coulmn

```python
model.fit(X_train, Y_train)
```

### **Check The Accuracy Score**
**Training Data**

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy Score On Traning Data ===>',training_data_accuracy)
```

**Testing Data**

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy Score On Testing Data ===>',test_data_accuracy)
```
